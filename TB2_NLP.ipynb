```python
# Install library yang dibutuhkan
!pip install nltk pandas

# Import library
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# Unggah file
from google.colab import files
uploaded = files.upload()

# Muat dataset
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)

# Tampilkan beberapa baris pertama untuk memahami struktur data
print("Pratinjau Dataset:")
print(data.head())

# Fungsi untuk menormalkan teks
def normalisasi_teks(teks):
    # Ubah menjadi huruf kecil
    teks = teks.lower()
    # Hapus angka
    teks = re.sub(r'\d+', '', teks)
    # Hapus tanda baca
    teks = re.sub(r'[^\w\s]', '', teks)
    # Hapus spasi ekstra
    teks = re.sub(r'\s+', ' ', teks).strip()
    return teks

# Pilih kolom yang akan diproses
kolom_yang_diproses = ['movie_title', 'genres', 'plot_keywords']
for kolom in kolom_yang_diproses:
    if kolom not in data.columns:
        print(f"Error: Kolom '{kolom}' tidak ditemukan dalam dataset.")
    else:
        # Normalisasi teks dalam kolom
        data[kolom] = data[kolom].astype(str).apply(normalisasi_teks)

        # Hapus stopwords
        stop_words = set(stopwords.words('english'))

        def hapus_stopwords(teks):
            kata_kata = nltk.word_tokenize(teks)
            kata_terfilter = [kata for kata in kata_kata if kata not in stop_words]
            return ' '.join(kata_terfilter)

        data[f'{kolom}_diproses'] = data[kolom].apply(hapus_stopwords)

# Simpan dataset yang telah diproses
nama_file_proses = 'processed_movie_metadata.csv'
data.to_csv(nama_file_proses, index=False)
print(f"Dataset yang telah diproses disimpan dengan nama {nama_file_proses}")
```

```python
# Import library yang diperlukan
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Unggah file CSV Anda di Colab, lalu gunakan file path sesuai dengan file yang diunggah.
# Jika Anda menggunakan file langsung, pastikan mengganti 'movie_metadata.csv' dengan nama file Anda.

# Langkah 1: Memuat file CSV
from google.colab import files
uploaded = files.upload()  # Anda dapat mengunggah file Anda di sini

# Pastikan nama file sesuai dengan yang diunggah
file_name = list(uploaded.keys())[0]  # Nama file pertama yang diunggah
data = pd.read_csv(file_name)

# Langkah 2: Pilih kolom yang ingin di-tokenisasi
columns_of_interest = ['movie_title', 'genres', 'plot_keywords']
selected_data = data[columns_of_interest]

# Langkah 3: Fungsi untuk tokenisasi
def tokenize_column(column):
    return column.fillna('').apply(lambda x: [token.strip() for token in x.split('|')] if isinstance(x, str) else [])

# Terapkan tokenisasi pada kolom yang dipilih
tokenized_data = selected_data.copy()
for column in columns_of_interest:
    tokenized_data[column] = tokenize_column(selected_data[column])

# Gabungkan kolom untuk analisis teks (misalnya, "plot_keywords" digunakan)
tokenized_data['combined_text'] = tokenized_data['plot_keywords'].apply(lambda x: ' '.join(x))

# Tambahkan kolom label (contoh: genre pertama sebagai label)
tokenized_data['label'] = tokenized_data['genres'].apply(lambda x: x[0] if len(x) > 0 else 'Unknown')

# Langkah 4: Bag of Words (BoW)
vectorizer_bow = CountVectorizer()
bow_matrix = vectorizer_bow.fit_transform(tokenized_data['combined_text'])

# Langkah 5: Split Data untuk Pelatihan dan Pengujian
X_train, X_test, y_train, y_test = train_test_split(bow_matrix, tokenized_data['label'], test_size=0.2, random_state=42)

# Langkah 6: Model Naive Bayes
model = MultinomialNB()
model.fit(X_train, y_train)

# Langkah 7: Prediksi dan Evaluasi
y_pred = model.predict(X_test)

# Hitung metrik evaluasi
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

# Tampilkan hasil evaluasi
print("\nMetrik Evaluasi:")
print(f"Akurasi: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Tampilkan laporan klasifikasi
unique_labels = sorted(set(y_test))  # Hanya gunakan label unik dari y_test
print("\nLaporan Klasifikasi:")
print(classification_report(y_test, y_pred, labels=unique_labels))

# Langkah 8: Simpan laporan ke file (opsional)
report_file = "classification_report.txt"
with open(report_file, "w") as f:
    f.write(classification_report(y_test, y_pred, labels=unique_labels))
print(f"\nLaporan klasifikasi disimpan dalam file: {report_file}")
```
